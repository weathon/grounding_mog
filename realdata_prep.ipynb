```python
import os
import cv2
```

```python
videos = os.listdir("/home/wg25r/Original_Videos")
```

```python
os.makedirs("/home/wg25r/original_frames", exist_ok=True)
```

```python
import tqdm

for video in videos:
    cap = cv2.VideoCapture(f"/home/wg25r/Original_Videos/{video}")
    os.makedirs(f"/home/wg25r/original_frames/{video}", exist_ok=True)
    i = 0
    bar = tqdm.tqdm(total=int(cap.get(cv2.CAP_PROP_FRAME_COUNT)))
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        cv2.imwrite(f"/home/wg25r/original_frames/{video}/{i:06d}.png", frame)
        i += 1
        bar.update(1)
    bar.close()
    cap.release()
    print(video, i)
```
